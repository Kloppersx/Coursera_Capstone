Loading some libraries and getting the dataframe from the wikipedia page

```python
import requests
import pandas as pd

#Get the table from the wiki webpage

url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
html = requests.get(url).content
df_list = pd.read_html(html)
df_wiki = df_list[0]

df_wiki.to_csv('my data.csv')

df_wiki.head()

#print(df_wiki.shape)
```

<h2>Part 1: Create a nice Dataframe</h2>
<h3>Getting some work done on the dataframe.</h3>

Lets remove all the rows where the field "Borough" is "Not assigned"

```python
#Remove the rows where Borough is Not assigned

df_wiki = df_wiki[df_wiki.Borough != 'Not assigned']

df_wiki = df_wiki.reset_index(drop=True)

df_wiki.head()
```

Where the neighbourhood is not assigned lets assign it the Borough name.

Even though there is only one instance where the neighbourhood is not assigned while the borough is, I preferred to
implement it in the most general way possible so it is potentially applicable to larger data sets.
Therefore I did a for loop that loops through all the rows and check if our case is satisfied or not.

```python
#Where the neighborhood is not assigned lets assign it the Borough name 

#print(len(df_wiki.index))

for k in range(0, len(df_wiki.index)):
    #print(k)
    if (df_wiki.iloc[[k]].Neighbourhood == 'Not assigned').bool():        
        #print (df_wiki.iloc[[k]].Neighborhood)
        #print (df_wiki.iloc[[k]].Borough)
        df_wiki.iloc[k, df_wiki.columns.get_loc('Neighbourhood')] = df_wiki.iloc[k, df_wiki.columns.get_loc('Borough')]

df_wiki.head(20)
#df_wiki.shape
```

Wonderful. Next step is to group the Neighbourhoods that presents the same Postcode and lets check if the dataframe built this way makes sense.


```python
#Lets group the Neighborhoods that presents the same Postcode. 

#HOLY COW THIS WAS HARD, TOOK ME ONE WHOLE AFTERNOON TO FIGURE OUT THE SINTAX (╯°□°)╯︵ ┻━┻

df_nice = df_wiki.groupby(["Postcode", "Borough"])['Neighbourhood'].apply(', '.join).reset_index()
df_nice                                                            
                                                                     
#┬─┬ノ( º _ ºノ) DONE
```

Its time to convert the Vanue Catogory Column in dummies varialbes (0,1) so we will be able to work with them later.

```python
Scarborough_onehot = pd.get_dummies(Scarborough_venues[['Venue Category']], prefix="", prefix_sep="")

Scarborough_onehot['Neighbourhood'] = Scarborough_venues['Neighbourhood'] 

fixed_columns = [Scarborough_onehot.columns[-1]] + list(Scarborough_onehot.columns[:-1])
Scarborough_onehot = Scarborough_onehot[fixed_columns]

Scarborough_onehot.head()
```

Lets get the mean frequency occurrance for each venues so we will be able to understand which are the most common venues 

```python
Scarborough_grouped = Scarborough_onehot.groupby('Neighbourhood').mean().reset_index()
Scarborough_grouped.head()
```

Now that we have the mean frequency we can print the 5 most common venues.

```python
num_top_venues = 5

for hood in Scarborough_grouped['Neighbourhood']:
    print("----"+hood+"----")
    temp = Scarborough_grouped[Scarborough_grouped['Neighbourhood'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

Just a function to ssort the venues in descending order.

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

Lets import the numpy library as np

```python
import numpy as np
```

Lets create a dataframe to display the to 5 most common venues

```python
num_top_venues = 5

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighbourhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighbourhood'] = Scarborough_grouped['Neighbourhood']

for ind in np.arange(Scarborough_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(Scarborough_grouped.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted.head(10)
```

Lets cluster the neighborhood in 5 clusters.

```python
kclusters = 5

Scarborough_grouped_clustering = Scarborough_grouped.drop('Neighbourhood', 1)

kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(Scarborough_grouped_clustering)

kmeans.labels_[0:10]
```

Now lets add the Cluster Label column in our dataframe.

```python
# add clustering labels
neighborhoods_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)

Scarborough_merged = Toronto_Scarborough_data

# merge toronto_grouped with toronto_data to add latitude/longitude for each neighborhood
Scarborough_merged = Scarborough_merged.join(neighborhoods_venues_sorted.set_index('Neighbourhood'), on='Neighbourhood')

Scarborough_merged.dropna(inplace=True)
Scarborough_merged = Scarborough_merged.astype({"Cluster Labels": int})
Scarborough_merged.head(100) # check the last columns!

#print(Scarborough_merged['Cluster Labels'].dtype)
```
